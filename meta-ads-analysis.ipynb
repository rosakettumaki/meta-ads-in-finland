```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("meta_ads_clean_csv.csv")
df.head()
```

```python
df.info()
df.describe()
df.columns
```

```python
df.sort_values(by="amount_spent", ascending=False).head(10)
```

```python
df.groupby("organisation_type")["amount_spent"].sum().reset_index().sort_values(by="amount_spent", ascending=False)
```

```python
summary = df.groupby("organisation_type")["amount_spent"].sum().reset_index()
summary.to_csv("total_by_org_type.csv", index=False)
```

```python
#TOTAL MONEY USED BYT ORGANISATION TYPE

summa = df.groupby("organisation_type")["amount_spent"].sum().reset_index()
summa.rename(columns={"amount_spent": "total_spending"}, inplace=True)
```

```python
#TOTAL NUMBER OF ADS BY ORGANISATION TYPE

maara = df.groupby("organisation_type")["number_of_ads"].sum().reset_index()
maara.rename(columns={"number_of_ads": "total_number_of_ads"}, inplace=True)
```

```python
yhdistetty = pd.merge(summa, maara, on="organisation_type")
yhdistetty.head()
```

```python
yhdistetty["ad_average_price"] = yhdistetty["total_spending"] / yhdistetty["total_number_of_ads"]
yhdistetty.head()
```

```python
yhdistetty.sort_values(by="ad_average_price", ascending=False)
```

```python
yhdistetty.sort_values(by="total_number_of_ads", ascending=False)
```

```python
#FINDING OUT THE NGO'S THAT SPENT THE MOST

ry_df = df[df["organisation_type"] == "Ry"]
```

```python
ry_top = ry_df.sort_values(by="amount_spent", ascending=False)
```

```python
ry_top[["page_name", "amount_spent", "number_of_ads"]].head(10)
```

```python
ry_top["price_per_ad"] = ry_top["amount_spent"] / ry_top["number_of_ads"]
ry_top[["page_name", "amount_spent", "number_of_ads", "price_per_ad"]].head(10)
ry_top = ry_top.sort_values(by="amount_spent", ascending=False)
ry_top.head(20)
```

```python
ry_top.to_csv("ry_top.csv", index=False)
```

```python
average_price_per_ad = ry_top["price_per_ad"].mean()
print(average_price_per_ad)
```
